## HPO란?

**하이퍼 파라미터**는 최적의 학습 모델을 구현하기 위해 사용자가 지정해주는 값

- 학습률(Learning rate), 배치 크기, 손실 함수 등 다양한 변수들이 하이퍼 파라미터에 해당
- 설정한 하이퍼 파라미터 값에 따라 모델의 성능이 달라지기 때문에 파라미터 값을 조정하여 최적의 조합을 찾는 것이 매우 중요(!)

이와 같은 하이퍼 파라미터를 최적화하는 문제를 Hyper Parameter Optimization, 줄여서 HPO라 합니다.

# HPO의 기본 방법은 무엇이 있을까요?

하이퍼 파라미터 최적화를 위해 결정되어야 하는 것은 아래와 같습니다. 아래 두 가지를 어떻게 설정하느냐에 따라 탐색 속도, 최적화 수준이 달라지게 됩니다.

- 하이퍼 파라미터 탐색 범위
- 하이퍼 파라미터 탐색 정책(방법)

아래에서는 HPO의 가장 대표적인 방법을 간단히 소개합니다.

## (1) Grid Search

image.png

- 특정 하이퍼 파라미터 구간에서 일정 간격으로 값을 선택하여, 선택된 모든 값을 탐색하는 최적해를 찾는 가장 기본적고 간단한 HPO 기법
- 구간 전역을 탐색하기 때문에 탐색 시간이 오래 걸리고, 균일한 간격으로 탐색하기 때문에 최적해를 찾지 못하는 경우가 발생할 수 있음

## (2) Random Search

image.png

- 탐색 구간 안에서 임의로 파라미터 값을 선택
- 그리드 서치의 문제점을 보완하기 위해 고안된 기법
- 구간 내 랜덤 조합을 사용하기 때문에 더 많은 지점을 살펴볼 수 있고 불필요한 반복 탐색이 줄어 그리드 서치보다 탐색 속도가 빠름

## (3) Bayesian Search

image.png

- 임의의 목적 함수를 최대로 하는 최적해를 찾는 기법. 이 때 목적함수는 손실함수와 같이 최적화하려는 대상을 의미
- 하이퍼 파라미터 값 선정 과정에서 사전지식을 반영하는 Sequential Model-based Global Optimization(SMBO) 방식
- 목적 함수의 형태에 대한 확률적 추정을 수행하는 **Surrogate** 모델과 Surrogate 모델 결과를 활용해 최적해 탐색에 유용한 후보를 추천하는 **Acquisition** 함수로 구성
