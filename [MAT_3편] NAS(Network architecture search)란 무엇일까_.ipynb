## NAS란?

NAS는 Network Architecture Search의 약어입니다. 여기서 Network는 신경망 모델을 의미하며, Network Architecture라 함은 신경망 모델의 구조를 의미합니다. 용어만 보면 간단히 신경망 모델 구조 탐색을 의미하는 것을 알 수 있습니다.

NAS가 무엇인지 알아보기에 앞서, 딥러닝 아키텍처(구조) 설계에서 몇 가지 고민을 살펴보겠습니다.

- 대부분  manual 하게 만들어짐 (해당 분야의 전문가 필요)

  - Ex) CNN에서 Hidden layer 수, layer당 filter 수, filter의 size, padding, stride, 언제 pooling 할 것인지?

- 어느 task에서 어떤 dataset을 쓰느냐에 따라 최적의 네트워크 구조 다름

NAS는 이런 고민으로부터 시작된 문제입니다. 이 문제를 해결는 다음의 목표를 달성하는 것입니다.

> __data__와 __task__를 대상으로 뉴런 연결 구조, 가중치를 탐색

## NAS의 구성요소

NAS는 아래 그림과 같이 표현을 할 수 있습니다. 그림에서 살펴보면, NAS는 크게 3가지 요소로 구성됩니다. 아래에서 그 3가지 요소인 Search Space, Search Strategy, Performance Estimation Strategy에 대해 알아보도록 하겠습니다.

 image.png

### 1. Search Space (탐색 범위)

- 알고리즘이 탐색을 수행하는 영역으로 layer수, Operation의 종류, 하이퍼 파라미터 등이 Search Space에 속함
- Search Strategy는 Search space에 있는 요소들로부터 후보 아키텍처를 선정함


### 2. Search Strategy (탐색 전략)

- Search Space의 요소들을 이용하여 최적 아키텍처를 찾기 위한 방법
- exploration(탐색 공간의 커버리지), exploitation(탐색 속도를 빠르게) 둘다 고려할 필요가 있음

### 3. Performance Estimation Strategy (모델 성능 추정 전략)

- Search Strategy에서 추출한 후보 아키텍처의 성능을 예측
- Search Strategy는 후보 아키텍처의 예상 성능을 반영하여 새로운 아키텍처를 만드는 과정을 반복

### 1. NAS with RL(Reinforcement Learning, 강화학습)


- 강화학습(RL)을 이용하여 NAS를 수행하는 연구가 진행됨

- 신경망 구조를 가변 길이 문자열로 정의함

  - 아래 그림에서 RNN의 출력 하나하나가 아키텍처의 하이퍼 파라미터

- RNN 컨트롤러(RL Agent)는 Validation Accuracy(RL reward)를 최대화하는 문자열을 탐색

### 2. AmebaNet

- 진화하는 방식을 취하는 Evolutionary Strategy를 이용

  - 랜덤하게 네트워크를 샘플링하고 개별 네트워크를 학습
  - 낮은 점수의 네트워크는 버리고 높은 점수의 네트워크는 복제하여 변이
  - 변이는 은닉층 연결을 변경하거나 Operation을 변경

### 3. DARTS

- Gradient Descent 방식을 도입

  - Weight 업데이트에 활용

- 셀 기반 Search Space를 사용
- Node를 연결하는 edge를 Operation 후보의 조합으로 간주
- Weight가 가장 큰 Operation을 선택
